```python
import numpy as np
import matplotlib.pyplot as plt

f = open("ans.txt", "r");
a=[]
b=[]
c=[]
k=0
for x in f:
        if(k%3==0):
                a.append(int(x)/1000000)
                k=k+1
        elif(k%3==1):
                b.append(int(x)/1000000)
                k=k+1
        elif(k%3==2):
                c.append(int(x)/1000000)
                k=k+1
f.close()
# for i in a:
#         print(i)
a0=[a[0],b[0],c[0]]
a1=[a[1],b[1],c[1]]
a2=[a[2],b[2],c[2]]
a3=[a[3],b[3],c[3]]
a4=[a[4],b[4],c[4]]
a5=[a[5],b[5],c[5]]
a6=[a[6],b[6],c[6]]
a7=[a[7],b[7],c[7]]
a8=[a[8],b[8],c[8]]
a9=[a[9],b[9],c[9]]
# creating the dataset
# set width of bar
barWidth = 0.05
fig = plt.subplots(figsize =(10, 5))

br1 = np.arange(len(a0))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]
br5 = [x + barWidth for x in br4]
br6 = [x + barWidth for x in br5]
br7 = [x + barWidth for x in br6]
br8 = [x + barWidth for x in br7]
br9 = [x + barWidth for x in br8]
br10 = [x + barWidth for x in br9]

# Make the plot
plt.bar(br1, a0, color ='g', width = barWidth, edgecolor ='grey')
plt.bar(br2, a1, color ='r', width = barWidth, edgecolor ='grey')
plt.bar(br3, a2, color ='g', width = barWidth, edgecolor ='grey')
plt.bar(br4, a3, color ='r', width = barWidth, edgecolor ='grey')
plt.bar(br5, a4, color ='g', width = barWidth, edgecolor ='grey')
plt.bar(br6, a5, color ='r', width = barWidth, edgecolor ='grey')
plt.bar(br7, a6, color ='g', width = barWidth, edgecolor ='grey')
plt.bar(br8, a7, color ='r', width = barWidth, edgecolor ='grey')
plt.bar(br9, a8, color ='g', width = barWidth, edgecolor ='grey')
plt.bar(br10, a9, color ='r', width = barWidth, edgecolor ='grey')

# Adding Xticks
plt.xlabel('Branch', fontweight ='bold', fontsize = 15)
plt.ylabel('Students passed', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(a0))],
		['SCHED_OTHER', 'SCHED_RR', 'SCHED_FIFO'])

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# set height of bar
IT = [12, 30, 1, 8, 22]
ECE = [28, 6, 16, 5, 10]
CSE = [29, 3, 24, 25, 17]

# Set position of bar on X axis
br1 = np.arange(len(IT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

# Make the plot
plt.bar(br1, IT, color ='r', width = barWidth, edgecolor ='grey', label ='IT')
plt.bar(br2, ECE, color ='g', width = barWidth, edgecolor ='grey', label ='ECE')
plt.bar(br3, CSE, color ='b', width = barWidth, edgecolor ='grey', label ='CSE')

# Adding Xticks
plt.xlabel('Branch', fontweight ='bold', fontsize = 15)
plt.ylabel('Students passed', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(ECE))],
		['2015', '2016', '2017', '2018', '2019'])

plt.legend()
plt.show()
```
